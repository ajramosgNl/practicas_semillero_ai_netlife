```python
import os
from openai import OpenAI

client = OpenAI(
    api_key="AIzaSyDjSbFVf25PmVRHiuqhF7Bd8mI5VaVaOI0", 
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

prompt = """
Soy un estudiante de ingenier√≠a sin experiencia en IA. Explica qu√© es machine learning en m√°ximo 3 l√≠neas, incluyendo 1 analog√≠a simple. Evita t√©rminos t√©cnicos complejos.
"""

top_p_values = [0.1, 0.3, 0.5, 0.7, 0.9, 1.0]

print("=== Probando Diferentes Valores de Top-P en Zero-Shot Prompting ===\n")

for top_p in top_p_values:
    print(f"\n--- Top-P: {top_p} (Diversidad: {'Muy Baja' if top_p < 0.3 else 'Baja' if top_p < 0.6 else 'Media' if top_p < 0.9 else 'Alta'}) ---")
    
    response = client.chat.completions.create(
        model="gemini-2.5-flash",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.7,  
        top_p=top_p,      
    )
    
    # Imprime la respuesta
    print(response.choices[0].message.content.strip())
    print("-" * 60)

print("\n=== Observaci√≥n Esperada ===")
print("- Top-P bajo (0.1-0.3): Respuestas muy similares, solo opciones 'seguras'.")
print("- Top-P medio (0.5-0.7): Variaci√≥n moderada en ejemplos y redacci√≥n.")
print("- Top-P alto (0.9-1.0): Respuestas m√°s diversas, m√°s creatividad en analog√≠as.")
print("\n=== Diferencia Clave ===")
print("Top-P controla Q√ö√ÅNTAS opciones considera el modelo.")
print("Con Top-P 0.1: 'Solo el 10% de opciones m√°s probables'")
print("Con Top-P 1.0: 'Considera todas las opciones posibles'")
```
